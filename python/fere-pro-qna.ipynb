```python
import asyncio
import websockets
import json
from uuid import UUID
import nest_asyncio

# Apply the nested event loop fix
nest_asyncio.apply()

async def stream_chat_response(user_id: UUID, query: str, base_url: str, api_key: str):
    """
    Opens a WebSocket connection to send a chat query and stream the response.

    Args:
        user_id (UUID): The user ID for the chat session.
        query (str): The query string to send.
        base_url (str): The base URL for the WebSocket endpoint.
        api_key (str): The API key for authentication.
    """
    # Construct the WebSocket URL
    ws_url = f"wss://{base_url}/chat/v2/ws/{user_id}?X-FRIDAY-KEY={api_key}"

    try:
        # Establish WebSocket connection
        async with websockets.connect(ws_url) as websocket:
            # Prepare the message payload
            message = {
                "message": query,
                "stream": True,
                "agent": "ProAgent"
            }

            # Send the message to the WebSocket
            await websocket.send(json.dumps(message))
            print("Message sent to WebSocket.")

            # Stream and display the response
            print("Streaming response:")
            async for response in websocket:
                print(response)
    except Exception as e:
        print(f"An error occurred: {e}")

```

```python
_user_id = "YOUR_USER_ID" # @param {type:"string"}
user_id = UUID(_user_id)
query = "Who are the top KOLs for $CHILLGUY?" # @param {type:"string"}
base_url = "api.fereai.xyz"  # Replace with the actual base URL
api_key = "YOUR_API_KEY"  # @param {type:"string"}

asyncio.run(stream_chat_response(user_id, query, base_url, api_key))
```
